```python
a = list(df.to_numpy())
l = []
for i in a:
    p = []
    i = list(i)
    for j in i:
        try:
            j = j.replace(',', '.')
            j = float(j)
        except:
            pass
        p.append(j)
    l.append(p)
l = np.array(l)
df = pd.DataFrame(l, columns = op)
for i in range(1, len(op)):
    df[op[i]] = df[op[i]].astype(float)
```

```python
df1 = pd.read_csv("кк_по_годам.csv")
```

```python
a = list(df1['Плотность населения'])
l = []
for i in a:
    try:
        i = i.replace(',', '.')
        i = float(i)
    except:
        pass
    l.append(i)
l = np.array(l)
df1['Плотность населения'] = np.array(l)
df1
```

```python
x = np.array(df1["Год"])
y = np.array(df1["Плотность населения"])
plt.plot(x, y)
plt.show()
```

```python
model = np.poly1d(np.polyfit(x, y, 4))
polyline = np.linspace(1991, 2022, 31)
plt.plot(x, y)
plt.plot(polyline, model(polyline))
plt.title("Сравнение реальных данных и предсказанных")
plt.xlabel("год")
plt.ylabel("плотность населения")
plt.show() 
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mpe
mpe(y, model(polyline))*100
```

```python
x = np.array(df1["Год"])
y = np.array(df1["Плотность населения"])
print(x)
print(y)
plt.plot(x, y)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


p = PolynomialFeatures(degree = 2, include_bias = False)


pf = p.fit_transform(x.reshape((-1, 1)))


prm = LinearRegression()
prm.fit(pf, y)


#print(poly_reg_model.intercept_, poly_reg_model.coef_)
```

```python
y_predicted = prm.predict(pf)

plt.xlabel("Год")
plt.ylabel("Плотность населения")
plt.plot(x, y)

plt.plot(x, y_predicted, color='purple')
```
